```python
# Load the data
df = pd.read_csv("diabetes.csv")

df.head(5)
```

```python
if df['Outcome'].dtype == object:
    df['Outcome'] = LabelEncoder().fit_transform(df['Outcome'])

print("Class distribution (0=Non-diabetic, 1=Diabetic):")
print(df['Outcome'].value_counts(normalize=True))

X = df.drop("Outcome", axis=1)
y = df["Outcome"]

# Split and scale
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Visualize first 2 features
plt.figure()
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train, cmap='bwr', edgecolors='k')
plt.title("2D Feature View of Training Set")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
